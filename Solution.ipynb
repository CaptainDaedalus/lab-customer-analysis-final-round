```python
import pandas as pd  # keywords are highlighted in green, other strings in red, etc.
import numpy as np
```

```python
#1. Show the DataFrame's shape.
mca = pd.read_csv('marketing_customer_analysis.csv')

display(mca.head())
display(mca.tail())
mca.shape
```

```python
# 2. Standardize header names.
#df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')

cols = []
for col in mca.columns:
    cols.append(col.lower())
mca.columns = cols

mca.head()
```

```python
# 3.Which columns are numerical?
num_col = mca._get_numeric_data()
display(num_col)
```

```python
#4. Which columns are categorical?
cat_col = mca.select_dtypes(include=['object']).columns
display(cat_col)
```

```python
#5. Check and deal with NaN values.
show_nan = mca[mca.isna().any(axis=1)] #using isna() (or isnull()) to select all rows with NaN under an entire DataFrame
mca_value = mca.isnull().values.any() #check for NaN values is True
mca_sum = mca.isnull().sum().sum() #check for NaN sum total is 9254
display(show_nan)
display(mca_value)
display(mca_sum)

```

```python
print(mca['vehicle type']) #this is the column with the NaN values
```

```python
#drop all rows that have NaN/None values

#Fill missing values of numeric columns with mean value

for col in mca.select_dtypes(include=np.number).columns:
   
    mca[col] = mca[col].fillna(np.mean(mca[col]))
```

```python
#Fill missing values of categorical columns with mode value
for col in mca.select_dtypes(include=['object']).columns:
    mca[col] = mca[col].fillna(mca[col].mode()[0])
```

```python
mca.isna().sum() #no more missing values
```

```python
#X-y split.
y = mca['total claim amount']
X = mca.drop(['total claim amount'], axis=1)

print(X.shape)
print(y.shape)
```

```python
##Normalize (numerical).
X_num = X.select_dtypes(np.number)
X_cat = X.select_dtypes(object)
```

```python
#Customer Analysis Round 6
#One Hot/Label Encoding (categorical).
#Concat DataFrames
#Linear Regression
#Train-test split.
#Apply linear regression.
#Model Validation
#Description:
#R2.
#MSE.
#RMSE.
#MAE.
```

```python
# We delete column 'customer' and 'effective_to_date' since it doesn't make sense to
# convert them to numerical.
X_cat_to_encode = X_cat.drop(['customer','effective to date'], axis=1)
X_cat_to_encode.columns.tolist()
```

```python
#One Hot/Label Encoding (categorical).

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
encoder.fit(X_cat_to_encode) # This determines how many unique values has each categorical column
print(encoder.categories_)
encoded = encoder.transform(X_cat_to_encode).toarray()
print(encoded)
onehot_encoded = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
onehot_encoded.head(20)
```

```python
# let's merge all this information together into a single dataset with all features, now numerical
X = pd.concat([pd.DataFrame(x_normalized, columns=X_num.columns), pd.DataFrame(onehot_encoded)],axis=1)
```

```python
# train test split is the way ML generates its claim to fame: 
# we build the model on a portion of the data but we then validate it in 
# another "fresh" portion
# our model has no opportunity to "cheat": it must accurately guess the values 
# in the "fresh" dataset that it never saw before
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=56)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Apply linear regression.
#we train/fit our model like yesterday
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred_train = lm.predict(X_train)
r2_score(y_train, y_pred_train)
```

```python
# But now we evaluate it in the TEST portion of the data, that we did not use for training.
# This way we know our model is genuinely guessing our donations, not just repeating the values it has seen in the training data


y_pred_test = lm.predict(X_test)
r2_score(y_test, y_pred_test)
```

```python
from sklearn.metrics import mean_squared_error
# Comptuing the RMSE -> root mean squared error.
np.sqrt(mean_squared_error(y_test,y_pred_test))
```
